```python
#Adding Fare Group to the dataset
Y = pd.read_csv("titanic.csv")
if "Fare" in Y.columns:
    Y['fare'] = pd.to_numeric(Y['Fare'], errors='coerce')
    fare_bins = [0, 20, 50, 100, 200, float('inf')]
    fare_labels = ['Low', 'Lower-Middle', 'middle', 'Upper-Middle', 'High' ]
    Y['Fare Group'] = pd.cut(Y['Fare']+1, bins=fare_bins, labels=fare_labels, include_lowest=True)
    print(Y[['PassengerId', 'Fare', 'Fare Group']].head())
    Y.to_csv('updated_titanic.csv', index=False)
else:
    print("NO 'Fare' column found in dataset")
```

```python
# extracting title from name
Y = pd.read_csv('titanic.csv')
Y['Title'] = Y['Name'].str.extract(r",\s(\w+)\.")
print("\nUnique titles in dataset")
print(Y["Title"].unique())
title_survival = Y.groupby("Title")["Survived"].mean()
print("\nSurvival rates by title:")
print(title_survival)
```
